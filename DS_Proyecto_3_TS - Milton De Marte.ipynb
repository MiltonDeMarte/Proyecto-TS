```python
#Unimos los 3 datasets, podemos hacerlo asi directamente ya que tienen las mismas columnas

dataset = pd.concat([data_2019, data_2018, data_2017])


#Nos quedamos solamente con la estación Illia y la forma de pago Efectivo

mask = np.logical_and(dataset.estacion=='Illia', dataset.forma_pago == 'EFECTIVO')
dataset = dataset[mask]


#Quitamos las columnas que no vamos a utilizar

dataset.drop(columns = ['periodo','hora_inicio','forma_pago', 'estacion'], inplace = True)

dataset.head()
```

```python
#Pasamos la columna fecha a formato datetime, y le unimos la hora_fin como hora

dataset['fecha2'] = pd.to_datetime(dataset.fecha) + pd.to_timedelta(dataset.hora_fin, unit = 'h')

#Quitamos las columnas fecha y hora_fin ya que las incluimos en el campo fecha2

dataset.drop(columns=['fecha','hora_fin'], inplace = True)

dataset.head()
```

```python
#Corregimos el nombre de fecha2 a fecha y de cantidad_pasos a cantidad_autos

dataset.rename(columns = {'fecha2':'fecha', "cantidad_pasos":"cantidad_autos"}, inplace = True) 
dataset.head()
```

```python
#Reordenamos el dataset por fecha

dataset.sort_values('fecha', inplace = True)
dataset.reset_index(drop=True, inplace = True)
dataset.head()
dataset.shape
```

```python
dataset.describe()
```

```python
#Graficamos nuestros datos, tal como están

plt.figure(figsize = (15,6))
plt.plot(dataset.fecha, dataset.cantidad_autos)
plt.title("Flujo de autos medido en Horas \n", color= 'r', fontsize= 15)
plt.xlabel('\n Tiempo', color= 'g', fontsize= 15)
plt.ylabel('\n Cantidad de autos, por hora', color= 'g', fontsize= 15)
plt.show()
```

### A.R.I.M.A

#### *p - d - q*
     
   * **p** = Es la "parte" auto-regresiva del modelo. Es la que permite incoporar la presencia de valores pasados al modelo. Ejemplificando: Seria igual a decir que es probable que mañana llueva, dado que hace 3 dias que esta lloviendo.
   
     
   * **d** = Es la "parte" integreada del modelo. Esto incluye los terminos en el modelo que incporporan el porcentaje de diferenciacion para aplicar a la series de tiempo. Ejemplificando: Seria igual a decir que, la temperatura de mañana es probable que sea igual a la de hoy si la variacion de la misma en los ultimos 3 dias fue muy poca.
     
     
   * **q** = Es la "parte" que representa la *media movil* del modelo. Esta permite presentar los errores del modelo como una respresentacion lineal de los errores observados de observaciones pasadas.
   
   

* _*Cuando trabajamos con series que tienen "efectos estacionales" usamos el modelo -seasonal A.R.I.M.A- la cual se puede representar como `ARIMA(p,d,q)(P,D,Q)`.*_


* _*Las primeras conotaciones `(p,d,q)` son los paramatros no-estacionales que fueron explicados arriba, mientras que `(P,D,Q)` siguen la misma definicion pero son aplicados a los parametros estacionales de la serie.*_

```python
# Se definen los parametros "p"-"d"-"q" para que tomen cualquier valor entre 0 y 2
p = q = range(4, 13)
d= range(0,3)

# Se generan todas las diferentes combinaciones
pdq = list(itertools.product(p, d, q))

# Se generan todas las diferentes combinaciones estacionarias
seasonal_pdq = [(x[0], x[1], x[2], 52) for x in list(itertools.product(p, d, q))]
```

* _*De esta forma, se puede hacer un "GridSearch" para buscar de forma iterativa las distinas combinaciones que puede usar -seasonal A.R.I.M.A-. Para cada combinacion de paramatros vamos a usar un nuevo modelo -seasonal A.R.I.M.A- con la funcion SARIMAX() de `statsmodel`.*_

```python
print("Distintas combinaciones de paramateros para -seasonal A.R.I.M.A")
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

* _*Una vez exploradas todas las posibilades de parametros, el set optimo de los mismos va a ser el que desempeñe el mejor resultado(de acuerdo a cada problematica). De esta manera esta manera, se puede usar las distintas combinaciones expuestas arriba para automatizar los procesos de entrenamiento y evaluacion.*_

#### *GridSearch y evaluacion*

* _*Al momento de evaluar y comparar modelos que fueron entrenados con distintos parametros, estos puede ser confrontar basandose en que tan bien se ajusto a los datos o sobre su habilidad para predecir con precision eventos futuros. En este caso voy a usar la metrica de evaluacion [AIC(Akaike Information Criterion)](https://www.sciencedirect.com/topics/medicine-and-dentistry/akaike-information-criterion#:~:text=Akaike%20information%20criterion%20(AIC)%20(,among%20all%20the%20other%20models.)(seccion 10.2.2). Esta misma mide que tan bien un modelo se ajusta a la informacion mientras toma en consideracion la complejidad general del modelo:*_

     * Un modelo que se ajusta bien sobre los datos mientras usa muchas features tendra un valor alto de indice AIC por sobre un modelo que usa menos features para lograr el mismo ajuste.
     
     
* _*Una puntuacion **baja** de AIC es en este caso en lo que estamos interesados.*_

```python
y= semanal.cantidad_autos

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model= sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results= model.fit()

            print('SARIMAX{}x{} - AIC:{}'.format(param, param_seasonal, results.aic))
            
        except:
            continue
```

* _*Itero sobre cada combinacion de parametros y uso la funcion SARIMAX() para ajustar los correspondientos modelos de -seasonal A.R.I.M.A-. En este caso el argumento "order" especifica los parametros para **p - d - q** mientras que el argumento "seasonal_order" especifica los componentes estacioneles **P - D - Q - S**(El termino **s** es la periocidad de la serie de tiempo, en este caso 12 para periodos anuales) del modelo -seasonal ARIMA-. Luego de ajustar cada modelo SARIMAX(), el codigo anterior devuelve los respectivos valores AIC para cada combinacion.*_


* _*Al observar los resultados, se observa que la combinacion de parametros `SARIMAX(1, 1, 1)x(0, 1, 1, 12)` con resultado **2491** es la correcta dado que es el menor de los valores AIC en los resultados.*_

#### *Entrenando*

```python
model= sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(0, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results= model.fit()

print(results.summary().tables[1])
```

* _*La tabla que devuelve los resultados de SARIMAX contiene mucha informacion importante. En este caso particular la columna **coef** resulta la de mayor interes. Esta misma represnta los pesos (importancia) de cada feature y como esta misma impacta en la serie.*_


* _*La columna P>|z| refiere al significado de cada peso de cada feature. Estos resultados son muy buenos dado que todos son menores a 0.05 lo cual significa que el modelo conserva importancia para todos.*_
