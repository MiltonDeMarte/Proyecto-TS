* *Extractos de la [nota](https://www.clarin.com/ciudades/crisis-llego-transito-cayo-cantidad-autos-ingresan-ciudad_0_KpXkY_Jqe.html) postulada:*


>*En lo que va del año, las naftas de mayor octanaje subieron un 68%.*

>*“El aumento del combustible es la principal razón por la que dejé de usar el auto”, dice Agustín Fuentes, un empleado administrativo de 31 años.*

>*En octubre, por los peajes de las autopista 25 de Mayo, Illia y Perito Moreno circularon 11.242.074 vehículos, un 3% menos que en octubre del año pasado.*

>*En las cabinas de la **Illia** hubo un 38% menos de tránsito, en las de Dellepiane cayó el 5%, y un 3% en las de Avellaneda.*

>*Los datos se desprenden de Autopistas Urbanas SA, la empresa estatal que administra estos caminos, y se complementan con análisis del INDEC. En cuestiones de circulación, el Instituto de Estadísticas y Censos viene mostrando indicadores negativos desde junio. El último dato disponible es de agosto y, en comparación al mismo mes de 2017, muestra una disminución del 5% en los accesos a la Ciudad (Norte, Oeste y Riccheri) y del 2% en las autopistas porteñas.*


* *Extractos de la [nota](https://www.perfil.com/noticias/sociedad/por-la-crisis-y-el-alza-de-nafta-y-peaje-entran-menos-autos-a-caba.phtml) postulada:*

>*De la mano de los ultimos aumentos de los precios de la nafta, del valor de los peajes para ingresar a la Ciudad mas las diferentes areas de restricciones de circulacion de vehiculos que se implementaron en la Capital Federal y las obras, cada vez menos autos transitan y avenidas porteñas. Al mismo tiempo -y a pesar de los aumentos en las tarifas, que se duplicaron en algunos casos-, aumentaron los usuarios de trenes, colectivos y subtes que circulan por la Ciudad.*

>*“Estamos percibiendo que hay menos autos, sobre todo en las horas pico. El tránsito se hace más fluido, por lo que se tarda menos tiempo en llegar a destino”, aseguró Fabián Pons, del Observatorio Vial Latinoamericano (Ovilam)[...] "Desde el último aumento de la nafta y los consejos que dio el Gobierno para ahorrar combustibles, se circula a una velocidad más moderada que meses atrás. Lo bautizamos como síndrome del taxista[...]*

>*La situación también se repite en los peajes de las autopistas del Acceso Norte, Oeste y Ricchieri. Durante septiembre de 2017 circularon unos 27,9 millones de vehículos mientras que en 2018 pasaron 25,9 millones, un 7% menos.*

>*Según datos aportados por la empresa Autopistas Urbanas SA (AUSA), durante el mes de septiembre ingresaron por las autopistas 25 de Mayo, Illia y Perito Moreno 10.014.357, un 7% menos de autos, si se los compara con el mismo mes del año pasado[...]se destacan los de Avellaneda y Dellepiane con el 8%; Alberti con el 4,5%, mientras que la Illia registró una caída del 3,2%.*
